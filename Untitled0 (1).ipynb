```python
# Install required packages
!pip install ultralytics opencv-python networkx scikit-image matplotlib


import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
from skimage.morphology import skeletonize
from ultralytics import YOLO

# 1️⃣ Load image and YOLO model
from google.colab import drive
drive.mount('/content/drive')

image_path = '/content/drive/MyDrive/output/test.jpg'   # top-view PCB image
model = YOLO('/content/drive/MyDrive/output/best.pt')  # your trained model

# Run detection
results = model(image_path, conf=0.4, iou=0.4, max_det=200)
boxes = results[0].boxes
xyxy = boxes.xyxy.cpu().numpy()
conf = boxes.conf.cpu().numpy()
cls = boxes.cls.cpu().numpy()
class_names = model.names

detections = pd.DataFrame({
    'xmin': xyxy[:,0], 'ymin': xyxy[:,1], 'xmax': xyxy[:,2], 'ymax': xyxy[:,3],
    'confidence': conf, 'class_id': cls.astype(int),
})
detections['name'] = detections['class_id'].apply(lambda x: class_names[x])

print("Detections:\n", detections)

# 2️⃣ Skeletonize traces
img = cv2.imread(image_path)
img = cv2.resize(img, (1600, 1200))
blur = cv2.GaussianBlur(img, (5,5), 0)
gray = cv2.cvtColor(blur, cv2.COLOR_BGR2GRAY)
th = cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                           cv2.THRESH_BINARY_INV,15,5)
edges = cv2.Canny(th,50,150)
binary = (edges>0).astype(np.uint8)
skeleton = skeletonize(binary)
skeleton_img = (skeleton*255).astype(np.uint8)
cv2.imwrite('skeleton.png', skeleton_img)

# 3️⃣ Build connectivity graph of traces
def build_graph(binary_img):
    G = nx.Graph()
    h, w = binary_img.shape
    for y in range(h):
        for x in range(w):
            if binary_img[y,x]:
                for dy in (-1,0,1):
                    for dx in (-1,0,1):
                        if dx==0 and dy==0: continue
                        ny, nx_ = y+dy, x+dx
                        if 0<=ny<h and 0<=nx_<w and binary_img[ny,nx_]:
                            G.add_edge((x,y),(nx_,ny))
    return G

G = build_graph(skeleton)

# 4️⃣ Estimate component pins from bounding boxes
def get_pins(xmin,ymin,xmax,ymax):
    w, h = xmax-xmin, ymax-ymin
    if w > h:
        p1 = (xmin, (ymin+ymax)//2)
        p2 = (xmax, (ymin+ymax)//2)
    else:
        p1 = ((xmin+xmax)//2, ymin)
        p2 = ((xmin+xmax)//2, ymax)
    return p1, p2

def nearest_node(G, pt):
    return min(G.nodes, key=lambda n: np.hypot(n[0]-pt[0], n[1]-pt[1]))

comp_pins = []
for i, r in detections.iterrows():
    p1, p2 = get_pins(r.xmin, r.ymin, r.xmax, r.ymax)
    n1, n2 = nearest_node(G, p1), nearest_node(G, p2)
    comp_pins.append({'idx': i, 'name': r.name, 'pin1':n1, 'pin2':n2})

# 5️⃣ Build netlist by grouping pins via connectivity
# Build a mapping from node to component pin IDs
node_to_pins = {}
for cp in comp_pins:
    for pin_name, node in [('pin1',cp['pin1']), ('pin2',cp['pin2'])]:
        node_to_pins.setdefault(node, []).append(f"{cp['name']}_{cp['idx']}_{pin_name}")

# Merge connected nodes into nets
visited = set()
nets = []
for node, pins in node_to_pins.items():
    if node in visited: continue
    net_nodes = list(nx.node_connected_component(G, node))
    visited |= set(net_nodes)
    net_pins = sum([node_to_pins.get(n, []) for n in net_nodes], [])
    if len(net_pins)>1:
        nets.append(net_pins)

print("\nNetlist (component_pin identifiers):")
for i, net in enumerate(nets):
    print(f"NET{i+1}: {net}")

# 6️⃣ Visualize simple schematic using NetworkX layout
H = nx.Graph()
for cp in comp_pins:
    H.add_node(f"{cp['name']}_{cp['idx']}_pin1", component=cp['name'])
    H.add_node(f"{cp['name']}_{cp['idx']}_pin2", component=cp['name'])
for net in nets:
    for i in range(len(net)-1):
        H.add_edge(net[i], net[i+1])

plt.figure(figsize=(10,10))
pos = nx.spring_layout(H, k=0.5)
nx.draw(H, pos, with_labels=True, node_size=400, font_size=8)
plt.title("PCB Schematic Approximation")
plt.show()

# 7️⃣ Save netlist as CSV
import csv
with open('netlist.csv', 'w') as f:
    w = csv.writer(f)
    w.writerow(['net_id','component_pins'])
    for i, net in enumerate(nets,1):
        w.writerow([f'NET{i}', ' '.join(net)])
print("\nNetlist saved to netlist.csv")
```
